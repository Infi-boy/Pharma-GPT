```python
# initialize driver
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import os
import pandas as pd
import re

# Setup Chrome options for headless mode
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
final = pd.read_csv('Data/final.csv', index_col = [0])
```

```python
index = pd.DataFrame()
def get_values(driver):
  smile, inchi_val, inchi_key = None, None, None
  links = driver.find_elements(By.TAG_NAME, 'a')
  # Loop through each <a> element and check if it contains the word "SMILES"
  for link in links:
      if smile != None and inchi_val != None and inchi_key != None:
        return smile, inchi_val, inchi_key
      if 'SMILES' == link.text:
        smile = re.search(
            r'>(.*?)<',
            str(
                link.find_element(
                    By.XPATH, "..").find_element(
                        By.XPATH, "..").find_element(
                            By.XPATH, "..").find_element(
                                By.TAG_NAME, 'ul').find_element(
                                    By.TAG_NAME, 'li').find_element(
                                        By.TAG_NAME, 'div'
                                        ).get_attribute('outerHTML')
                                        )
            ).group(1)
      if 'InChI' == link.text:
        inchi = link.find_element(
            By.XPATH, '..'
            ).find_element(
                By.XPATH, '..'
                ).find_element(
                    By.XPATH, '..'
                    ).find_element(
                        By.TAG_NAME,'ul'
                        ).find_elements(
                            By.TAG_NAME,'li'
                            )
        if len(inchi) == 2:
          inchi_val = re.search(
              r'>(.*?)<',
              str(
                  inchi[0].find_element(
                      By.TAG_NAME,'div'
                      ).get_attribute(
                          'outerHTML'
                          )
                      )
              ).group(1)
          inchi_key = re.search(
              r'>(.*?)<', str(
                  inchi[1].find_element(
                      By.TAG_NAME,'div'
                      ).get_attribute(
                          'outerHTML'
                          )
                      )
              ).group(1)
        elif len(inchi) == 1:
          inchi_val = re.search(
              r'>(.*?)<',
              str(
                  inchi[0].find_element(
                      By.TAG_NAME,'div'
                      ).get_attribute(
                          'outerHTML'
                          )
                      )
              ).group(1)
  return smile, inchi_val, inchi_key

for i in final.compound.unique():
  dic = {}
  index = pd.DataFrame()
  driver.get('https://en.wikipedia.org/wiki/' + i)
  try:
    dic[i] = list(get_values(driver))
  except:
    continue
  index = pd.concat(
      [
          index,
          pd.DataFrame(dic).T.rename(
              columns = {
              0:'SMILE',
              1:'InChI',
              2:'InchI_key'
              }
              )
          ]
      )
  path = 'Data/wiki_index.csv'
  if os.path.exists(path):
    index.to_csv(path, mode = 'a', header = False)
  else:
    index.to_csv(path)
  print(i)
```
